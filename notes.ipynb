```python
# Skewness using matplotlib 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, mode

# 1. Generate sample skewed data (e.g., positively skewed exponential data)
np.random.seed(42)
data = np.random.exponential(scale=3, size=1000)

# 2. Calculate summary statistics and skewness score
data_mean = np.mean(data)
data_median = np.median(data)
# Round values to find a clear peak for the mode if continuous
data_mode = mode(np.round(data, 1), keepdims=True)[0][0]
skewness_val = skew(data)

# 3. Initialize the Matplotlib figure
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")

# 4. Plot distribution using Seaborn
# histplot handles both the bars and the smooth KDE curve line
sns.histplot(data, kde=True, color="#4C72B0", stat="density", alpha=0.4)



# 6. Finalize graph aesthetics and text indicators
plt.title(f"Data Distribution (Calculated Skewness: {skewness_val:.2f})", fontsize=14, pad=15)
plt.xlabel("Values", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.legend(fontsize=11)

# Render the layout properly
plt.tight_layout()
plt.show()















```
